```python
import h5py
import json

# --- Set your H5 file name here ---
filename = '/content/model.h5'

try:
    with h5py.File(filename, 'r') as f:
        # Load the model's configuration
        model_config_str = f.attrs.get('model_config')
        if not model_config_str:
            raise ValueError("Model config not found in H5 file.")

        # Convert the JSON string to a Python dictionary
        model_config = json.loads(model_config_str)

        # Print the entire architecture for review
        print("--- Full Model Architecture ---")
        print(json.dumps(model_config, indent=2))
        print("\n" + "="*30 + "\n")

        # Extract and print the input shape
        try:
            # The first layer's config contains the input shape
            first_layer_config = model_config['config']['layers'][0]['config']
            input_shape = first_layer_config.get('batch_input_shape')
            if input_shape:
                 # We only need the shape, not the batch size (None)
                print(f"✅ Found Input Shape: {input_shape[1:]}")
            else:
                print("Could not automatically determine input shape from the first layer.")
        except (KeyError, IndexError):
            print("Could not find the input layer configuration.")

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
%%shell
eval "$(conda shell.bash hook)"
conda activate myenv

# This command creates a dummy my_model.h5 file for demonstration.
# The architecture here MUST match the one defined in the python script below.
python -c "import numpy as np; import tensorflow as tf; from tensorflow.keras.models import Sequential; from tensorflow.keras.layers import Dense, InputLayer; model = Sequential([InputLayer(input_shape=(13,)), Dense(16, activation='relu', kernel_initializer='he_normal'), Dense(37, activation='sigmoid')]); model.compile(optimizer='adam', loss='binary_crossentropy'); model.save('my_model.h5')"


python << END

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer
from tensorflow.python.framework import graph_io
from tensorflow.keras import backend as K

# Disable eager execution for TF 1.15
tf.compat.v1.disable_eager_execution()

def freeze_graph(model, session, output_node_names, output_filename):
    """
    Freezes a Keras model and saves it as a TensorFlow protobuf file.
    """
    graph_def = session.graph.as_graph_def()
    frozen_graph = tf.compat.v1.graph_util.convert_variables_to_constants(
        session,
        graph_def,
        output_node_names
    )
    with tf.io.gfile.GFile(output_filename, "wb") as f:
        f.write(frozen_graph.SerializeToString())
    print(f"Frozen graph saved to {output_filename}")


# --- 1. DEFINE YOUR FINAL MODEL ARCHITECTURE HERE ---
#
# This architecture is based on the JSON output from your .h5 file.
#
model = Sequential([
    # Input shape is (13,) based on "batch_shape": [null, 13]
    InputLayer(input_shape=(13,)),

    # First hidden layer with 16 units and relu activation.
    Dense(16, activation='relu', kernel_initializer='he_normal'),

    # Output layer with 37 units and sigmoid activation.
    Dense(37, activation='sigmoid')
])
#
# --- End of model definition ---


# 2. Load the weights into your defined model structure
model.load_weights('/content/model.h5')
print("Successfully loaded weights into the model.")

# 3. Freeze the graph
K.set_learning_phase(0)
output_names = [out.op.name for out in model.outputs]
sess = K.get_session()
freeze_graph(model, sess, output_names, "/content/frozen_model.pb")

END
```
